1- Chip de 32x16 com 4 bancos 

Criar um chip de 32x16 com 4 bancos

```python
# ==========================================================
# CHIP DE MEMÓRIA 32x16 COM 8 BANCOS
# ==========================================================

class ChipMemoria:

    def __init__(self):
        # Número de bancos
        self.bancos = 8

        # Cada banco possui 32 endereços
        self.enderecos = 32

        # Cada posição armazena 16 bits
        # Inicialmente todas recebem 0
        self.memoria = [
            [0 for _ in range(self.enderecos)]
            for _ in range(self.bancos)
        ]

    # ======================================================
    # Escreve um valor em um banco e endereço
    # ======================================================
    def escrever(self, banco, endereco, valor):

        # Verifica se o banco existe
        if not (0 <= banco < self.bancos):
            print("Banco inválido!")
            return

        # Verifica se o endereço existe
        if not (0 <= endereco < self.enderecos):
            print("Endereço inválido!")
            return

        # Garante que apenas 16 bits sejam armazenados
        self.memoria[banco][endereco] = valor & 0xFFFF

        print(f"Gravado {valor:04X} no Banco {banco}, Endereço {endereco}")

    # ======================================================
    # Lê um valor da memória
    # ======================================================
    def ler(self, banco, endereco):

        if not (0 <= banco < self.bancos):
            print("Banco inválido!")
            return

        if not (0 <= endereco < self.enderecos):
            print("Endereço inválido!")
            return

        valor = self.memoria[banco][endereco]

        print(f"Banco {banco} | Endereço {endereco:02d} = {valor:04X}")

        return valor

    # ======================================================
    # Mostra um banco inteiro
    # ======================================================
    def mostrar_banco(self, banco):

        if not (0 <= banco < self.bancos):
            print("Banco inválido!")
            return

        print(f"\n===== BANCO {banco} =====")

        for endereco, valor in enumerate(self.memoria[banco]):
            print(f"{endereco:02d} -> {valor:04X}")

        print("=========================")


# ==========================================================
# Programa Principal
# ==========================================================

def main():

    chip = ChipMemoria()

    while True:

        print("\n===== CHIP 32x16 =====")
        print("1 - Escrever")
        print("2 - Ler")
        print("3 - Mostrar Banco")
        print("4 - Sair")

        opcao = input("Escolha: ")

        if opcao == "1":

            banco = int(input("Banco (0-7): "))
            endereco = int(input("Endereço (0-31): "))
            valor = int(input("Valor (hex): "), 16)

            chip.escrever(banco, endereco, valor)

        elif opcao == "2":

            banco = int(input("Banco (0-7): "))
            endereco = int(input("Endereço (0-31): "))

            chip.ler(banco, endereco)

        elif opcao == "3":

            banco = int(input("Banco (0-7): "))

            chip.mostrar_banco(banco)

        elif opcao == "4":

            print("Programa encerrado.")
            break

        else:
            print("Opção inválida!")


if __name__ == "__main__":
    main()
```

Projete um sistema de memória que possui 16 linhas de dados (BD). Os pinos de endereço precisam ser menores que 14. A memória deve ter capacidade de 32 Gbits. Justifique cada decisão do projeto. Desenhe o chip. Qual é o tamanho do barramento de endereços?


Passo 1 – Definir o barramento de endereços
O enunciado diz que ele deve ser menor que 14.
O maior valor possível é:
13" bits"

Logo,
Barramento de Endereços = 13 bits

Isso fornece:
2^13=8192 endereços.

________________________________________
Passo 2 – Calcular o tamanho da palavra
A capacidade é:
32" Gbits"=32×2^30= 2^35 " bits"

Como existem apenas 2^13 endereços:

2^35/2^13 =2^22=4" " 194" " 304" bits"

Cada endereço armazena 4.194.304 bits.
Logo,
Palavra = 4.194.304 bits
________________________________________
Passo 3 – Verificar o barramento de dados
O BD possui apenas 16 bits.

Isso significa que uma palavra inteira não pode ser transferida em uma única operação.

Quantidade de transferências necessárias:
2^22/16=2^18=262" " 144

Ou seja, são necessárias 262.144 transferências para ler ou gravar uma palavra completa.

Justificativas
Escolha do barramento de endereços

Foi escolhido 13 bits porque o enunciado exige um valor inferior a 14 bits. Com isso, obtêm-se 8192 endereços.
________________________________________
Escolha da palavra

Como a memória possui 32 Gbits e apenas 8192 endereços, cada endereço deve armazenar:

4" " 194" " 304" bits"

Essa é a largura da palavra.
________________________________________
Barramento de dados

O barramento é fixado em 16 bits pelo enunciado.

Assim, cada acesso transfere somente 16 bits da palavra.
